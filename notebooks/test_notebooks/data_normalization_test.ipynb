## 2. write the data (only numbers and "Healthy" tag)

```python
import numpy as np
import pandas as pd
```

```python
title = []

Calories = []
Carbohydrates = []
Fat = []
Protein = []
Sodium = []
Fiber = []
Saturated_Fat = []
Polyunsaturated_Fat = []
Monounsaturated_Fat = []
Cholesterol = []

healthy = []

rating = []
review_count = []
make_again = []
```

```python
#for i in range(1000):
    
#    r = recipe_list[i]
for r in recipe_list:
    
    try:
        title.append(r.get_title())
    except:
        continue
    
    nu = r.get_nutritions()
    #print(nu)
    
    try:
        Calories.append(nu['Calories'])
    except:
        Calories.append(None)        
    try:
        Carbohydrates.append(nu['Carbohydrates'])
    except:
        Carbohydrates.append(None)        
    try:
        Fat.append(nu['Fat'])
    except:
        Fat.append(None)        
    try:
        Protein.append(nu['Protein'])
    except:
        Protein.append(None)        
    try:
        Sodium.append(nu['Sodium'])       
    except:
        Sodium.append(None)        
    try:
        Fiber.append(nu['Fiber'])
    except:
        Fiber.append(None)        
    try:
        Saturated_Fat.append(nu['Saturated Fat'])
    except:
        Saturated_Fat.append(None)        
    try:
        Polyunsaturated_Fat.append(nu['Polyunsaturated Fat'])
    except:
        Polyunsaturated_Fat.append(None)       
    try:
        Monounsaturated_Fat.append(nu['Monounsaturated Fat'])
    except:
        Monounsaturated_Fat.append(None)        
    try:
        Cholesterol.append(nu['Cholesterol'])
    except:
        Cholesterol.append(None)
    
    if 'Healthy' in r.get_tags():
        healthy.append(1)
    else:
        healthy.append(0)

    try:
        rating.append(r.get_rating())
    except:
        rating.append(None)
    try:
        review_count.append(r.get_reviewcount())
    except:
        review_count.append(None)
    try:
        make_again.append(r.get_makeagain())
    except:
        make_again.append(None)
```

```python
len(title)
```

```python
sum(healthy)
```

```python
df = pd.DataFrame({'title':title, 'healthy':healthy, 'rating':rating, 'review_count':review_count, 
                   'make_again': make_again, 'Calories':Calories, 'Carbohydrates':Carbohydrates, 'Fat':Fat, 
                   'Protein':Protein, 'Sodium':Sodium, 'Fiber':Fiber, 'Cholesterol':Cholesterol})
df.head()
```

## 3. try to load data from json file: "data/epi-recipes-json.json"

```python
from pandas.io.json import json_normalize

df3 = pd.DataFrame()
count = 1

for line in open('../data/epi-recipes-json.json','r'):
    count += 1
    data = json.loads(line)
    df3 = df3.append(json_normalize(data))
    if count%5000 == 0:
        print(count)
```
