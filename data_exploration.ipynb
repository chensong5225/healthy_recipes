```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rawdf = pd.read_csv('data/epi_r.csv')
rawdf.describe()
```

There are more than 20000 recipes in the dataset. Loading log of graphlab said there are 4-5 rows failed to parse.

```python
df = rawdf.dropna()
df.describe()
```

After remove those have no value for nutrition facts, there are 15864 recipes.

```python
print(df['healthy'].value_counts()/df['healthy'].count())
print(df['healthy'].value_counts())
```

Among the 15864 recipes there are 1264 recipes marked as healthy, which is about 8%. We can use a scatter matrix plot to see the relationship between the numbers.

```python
numdf = df[['title','rating','calories','protein','fat','sodium']]
numdf.describe()
```

```python
colors = []
palette = {0:'red', 1:'green'}
for c in np.nditer(df.healthy):
    colors.append(palette[int(c)])

scatterplot = pd.scatter_matrix(numdf, alpha = 0.3, figsize = (10,10), 
                                diagonal = 'hist', c = colors, marker = 'o', grid = True)
```
